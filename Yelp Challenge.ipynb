```python
import json
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
```

```python
neighborhoods = {}
cities = []
states = []
postal_codes = []
latitudes = {}
longitudes = {}
stars = {}
review_count = {}
alcohol = {}
bike_parking = {}
accepts_credit_cards = {}
caters = {}
drivethru = {}
goodforkids = {}
hastv = {}
noise_level = {}
outdoor_seating = {}
restuarants_price_range = {} 
delivery = {}
goodforgroups = {}
pricerange = {}
reservations = {}
table_service = {}
takeout = {}
wifi = {}
```

```python
business_mappings = {}
with open('business.json','r') as f:
    for i, line in enumerate(f):
        d = json.loads(line)
        attributes = d['attributes']
        if attributes != None:
            business_id = d['business_id']
            city = d['city']
            state = d['state']
            cities[business_id] = city
            states[business_id] = state

            postal_codes[business_id] = d['postal_code']
            latitude = d['latitude']
            if latitude == None:
                latitude = -1
            latitudes[business_id] = latitude

            longitude = d['longitude']
            if longitude == None:
                longitude = -1
            longitudes[business_id] = longitude
            star_rating = float(d['stars'])
            stars[business_id] = star_rating

            review_count[business_id] = float(d['review_count'])
            
            if "Alcohol" in attributes:
                alcohol[business_id] = attributes["Alcohol"]
            else:
                alcohol[business_id] = -1
            if "BikeParking" in attributes:
                bike_parking[business_id] = attributes["BikeParking"]
            else:
                bike_parking[business_id] = -1
            if "BusinessAcceptsCreditCards" in attributes:
                accepts_credit_cards[business_id] = attributes["BusinessAcceptsCreditCards"]
            else:
                accepts_credit_cards[business_id] = -1
            if "Caters" in attributes:
                caters[business_id] = attributes["Caters"]
            else:
                caters[business_id] = -1
            if "GoodForKids" in attributes:
                goodforkids[business_id] = attributes["GoodForKids"]
            else:
                goodforkids[business_id] = -1
            if "DriveThru" in attributes:
                drivethru[business_id] = attributes["DriveThru"]
            else:
                drivethru[business_id] = -1
            if "HasTV" in attributes:
                hastv[business_id] = attributes["HasTV"]
            else:
                hastv[business_id] = -1
            if "NoiseLevel" in attributes:
                noise_level[business_id] = attributes["NoiseLevel"]
            else:
                noise_level[business_id] = -1
            if "OutdoorSeating" in attributes:
                outdoor_seating[business_id] = attributes["OutdoorSeating"]
            else:
                outdoor_seating[business_id] = -1
            if "RestaurantPriceRange" in attributes:
                restuarants_price_range[business_id] = attributes["RestaurantPriceRange"]
            else:
                restuarants_price_range[business_id] = -1
            if "RestaurantsDelivery" in attributes:
                delivery[business_id] = attributes["RestaurantsDelivery"]
            else:
                delivery[business_id] = -1
            if "RestaurantsGoodForGroups" in attributes:
                goodforgroups[business_id] = attributes["RestaurantsGoodForGroups"]
            else:
                goodforgroups[business_id] = -1
            if "RestaurantsPriceRange2" in attributes:
                pricerange[business_id] = attributes["RestaurantsPriceRange2"]
            else:
                pricerange[business_id] = -1
            if "RestaurantsReservations" in attributes:
                reservations[business_id] = attributes["RestaurantsReservations"]
            else:
                reservations[business_id] = -1
            if "RestaurantsTableService" in attributes:
                table_service[business_id] = attributes["RestaurantsTableService"]
            else:
                table_service[business_id] = -1
            if "RestaurantsTakeOut" in attributes:
                takeout[business_id] = attributes["RestaurantsTakeOut"]
            else:
                takeout[business_id] = -1
            if "WiFi" in attributes:
                wifi[business_id] = attributes["WiFi"]
            else:
                wifi[business_id] = -1
```

```python
categoriesHashmap = {}
with open('business.json','r') as f:
    for i,line in enumerate(f):
        d = json.loads(line)
        if (d['categories'] != None):
            wordList = d['categories'].split(', ')
            for i in wordList:
                if i not in categoriesHashmap:
                    categoriesHashmap[i]= 1
                else:
                    categoriesHashmap[i]+=1
print(categoriesHashmap)
```

```python
categoriesHashmap = sorted(categoriesHashmap.items(), key = lambda kv:(kv[1], kv[0]), reverse = True)
```

```python
categoriesHashmap
```

```python
with open('categoryDict.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Category', 'Number'])
    for key, value in categoriesHashmap:
        writer.writerow([key, value])
```
